```python
#we use groupby bivarient and sum it then sort in descending order and convert entire thing into dataframe(reset_index)
df_cat_installs=df.groupby(['Category'])['Installs'].sum().sort_values(ascending=False).reset_index()
```

#### what are the Top 5 most installed Apps in each popular categories??

```python
df.head()
```

```python
dfa=df.groupby(['Category','App'])['Installs'].sum().reset_index()
```

```python
dfa
```

```python
dfa.sort_values('Installs',ascending=False)
```

```python
dfa[dfa.Category=='GAME'][:5]
```

```python
dfa[dfa.Category=='PRODUCTIVITY'][:5]
```

```python
dfa[dfa.Category=='FAMILY'][:5]
```

```python
apps=['GAME','COMMUNICATION','PRODUCTIVITY','SOCIAL']
sns.set_context('poster')
sns.set_style('darkgrid')

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):  #enumerate means giving the index 0-game 1-communication 2-productivity 3-social
    df2=dfa[dfa.Category==app]
    df3=df2.head()
    plt.subplot(4,2,i+1) size # size of plot h,w and no. of plot 0,1,2,3
    sns.barplot(x='Installs',y='App', data=df3)
    plt.xlabel('Insatllation in Millions')
    plt.ylabel('')
    plt.title(app,size=20)
    
plt.tight_layout() #for adjustment 
plt.subplots_adjust(hspace= .3)
plt.show()
```
